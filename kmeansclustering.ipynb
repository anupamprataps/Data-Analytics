```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df=pd.read_csv('CLV.csv')
df.head()
x=df.iloc[:,[0,1]]
km=KMeans(n_clusters=4,init='k-means++', max_iter=300, n_init=10, random_state=0)
km.fit(x)
km.fit(x).score(x)
df['cluster']=km.predict(x)
x1=df[df['cluster']==0]
x2=df[df['cluster']==1]
x3=df[df['cluster']==2]
x4=df[df['cluster']==3]
ctr=pd.DataFrame(km.cluster_centers_)
plt.scatter(x1['INCOME'],x1['SPEND'],c='red')
plt.scatter(x2['INCOME'],x2['SPEND'],c='blue')
plt.scatter(x3['INCOME'],x3['SPEND'],c='green')
plt.scatter(x4['INCOME'],x4['SPEND'],c='yellow')
plt.scatter(ctr[0],ctr[1],c='violet')
plt.show()
```
